# Listas e tuplas

```python
##CRIAÇÃO  E INICIALIZAÇÃO DE UMA LISTA PARA ARMAZENAR LEITURAS DE TEMPERATURAS.
temperatura_celsius = [25.1, 25.5, 24.9, 26.0, 25.8]
print(f"Leituras iniciais de temperatura: {temperatura_celsius}")
```

```python
##adcionando novas leituras á lista
###função .append()
temperatura_celsius.append(26.2)
temperatura_celsius.append(25.7)
print(f"leitura após novas coletas: {temperatura_celsius}")
```

```python
##3.acessando elementos especificos
primeira_leitura = temperatura_celsius[0]
ultima_leitura = temperatura_celsius[-1]#acessando ultimo elemento
print(f"primeira leitura registrada:{primeira_leitura} °C")
print(f"ultima leitura registrada:{ultima_leitura} °C")
```

```python
#4modificando leitura
temperatura_celsius[2]=25.0 #supondo que a terceira leitura (indice 2) foi corrigida
print(f"leituras após correção: {temperatura_celsius}")
```

```python
##5.Removendo uma leitura
temperatura_celsius.pop(0) #Remove o primeiro elemento(leitura mais antiga)
print (f"Leituras após remover a mais antiga: {temperatura_celsius}")
```

```python
##6.Verificando o tamanho da lista
print (f"Número total de leituras registradas: {len(temperatura_celsius)}")
```

```python
##7.verificando a existencia de um valor
if 26.2 in temperatura_celsius:
  print ("A temperatura de 26.2°c está registrada")
```

```python
##8.verificando a existencia e posição de um valor
temperatura_para_buscar = 26.2
if temperatura_para_buscar in temperatura_celsius:
  posicao = temperatura_celsius.index(temperatura_para_buscar)
  print(f"A temperatura de {temperatura_para_buscar}°C existe e está na posição (índice) {posicao} da lista.")
else:
  print(f"A temperatura de {temperatura_para_buscar}°C não está registrada na lista.")
```

```python
##8.ordenando as leituras
temperatura_celsius.sort()
print(f"leituras ordenadas:{temperatura_celsius}")
```

# Dicionarios

```python
##Criação e inicialização de um dicionário para armazenar estados de máquinas
estados_maquina ={
    "maquina_1": "operacional",
    "maquina_2": "manutenção",
    "maquina_3": "parada",
    "maquina_4": "operacional"
}
print(f"Estados iniciais de máquinas: {estados_maquina}")
```

```python
##adicionando novos estado áe  maquinas
estados_maquina.update({"maquina_5": "operacional"})
estados_maquina.update({"maquina_6": "manutenção"})
print(f"Estados após adcionar novas maquinas: {estados_maquina}")
```

```python
##Acessando estados especificos
estado_maquina_1= estados_maquina["maquina_1"]
estado_maquina_3 = estados_maquina["maquina_3"]
print(f"Estado da maquina 1: {estado_maquina_1}")
print(f"Estado da maquina 3: {estado_maquina_3}")
```

```python
##modificando um estado
estados_maquina["maquina_2"]= "operacional" #supondo que voltou a operacionar
print(f"Estados após atualização: {estados_maquina}")
```

```python
##Removendo uma leitura:
del estados_maquina["maquina_3"] #Remove o estado da maquina 3 ('desativada')
print(f"Estados apos a remoção da do estado da maquina 3: {estados_maquina}")
```

```python
## Apagando o estado de uma máquina (sem removê-la)
estados_maquina["maquina_1"] = None # Define o estado da maquina_1 como None
print(f"Estados após 'apagar' o estado da máquina 1: {estados_maquina}")

# Outra opção seria usar uma string vazia:
# estados_maquina["maquina_2"] = ""
# print(f"Estados após 'apagar' o estado da máquina 2 com string vazia: {estados_maquina}")
```

```python
##verificando o tamanho do dicionario:
print(f"Número total de estados registrados: {len(estados_maquina)}")
```

```python
##listar todas as maquinas cadastradas
print(f"Máquinas atualmente registradas: {estados_maquina.keys()}")
```

```python
##definindo um dicionario de comandos para as maquinas:
comandos_maquinas={
    "maquina_A": ["ligar", "monitorar","desligar"],
    "maquina_B": ["manutencao", "calibrar"],
    "maquina_c":["ligar","monitorar"],
    "maquina_D":["desligar","ligar"]
}
print(f"\nComandos para as máquinas: {comandos_maquinas}")
```

```python
##Executando os proximos comandos para a maquina A
proximo_comando = comandos_maquinas['maquina_A'].pop(0)
print(f"Executando comando: {proximo_comando} para a maquina A")
print (f"Comandos restantes para a maquina A: {comandos_maquinas['maquina_A']}")
```
