# Listas e tuplas

```python
##CRIAÇÃO  E INICIALIZAÇÃO DE UMA LISTA PARA ARMAZENAR LEITURAS DE TEMPERATURAS.
temperatura_celsius = [25.1, 25.5, 24.9, 26.0, 25.8]
print(f"Leituras iniciais de temperatura: {temperatura_celsius}")
```

```python
##adcionando novas leituras á lista
###função .append()
temperatura_celsius.append(26.2)
temperatura_celsius.append(25.7)
print(f"leitura após novas coletas: {temperatura_celsius}")
```

```python
##3.acessando elementos especificos
primeira_leitura = temperatura_celsius[0]
ultima_leitura = temperatura_celsius[-1]#acessando ultimo elemento
print(f"primeira leitura registrada:{primeira_leitura} °C")
print(f"ultima leitura registrada:{ultima_leitura} °C")
```

```python
#4modificando leitura
temperatura_celsius[2]=25.0 #supondo que a terceira leitura (indice 2) foi corrigida
print(f"leituras após correção: {temperatura_celsius}")
```

```python
##5.Removendo uma leitura
temperatura_celsius.pop(0) #Remove o primeiro elemento(leitura mais antiga)
print (f"Leituras após remover a mais antiga: {temperatura_celsius}")
```

```python
##6.Verificando o tamanho da lista
print (f"Número total de leituras registradas: {len(temperatura_celsius)}")
```

```python
##7.verificando a existencia de um valor
if 26.2 in temperatura_celsius:
  print ("A temperatura de 26.2°c está registrada")
```

```python
##8.verificando a existencia e posição de um valor
temperatura_para_buscar = 26.2
if temperatura_para_buscar in temperatura_celsius:
  posicao = temperatura_celsius.index(temperatura_para_buscar)
  print(f"A temperatura de {temperatura_para_buscar}°C existe e está na posição (índice) {posicao} da lista.")
else:
  print(f"A temperatura de {temperatura_para_buscar}°C não está registrada na lista.")
```

```python
##8.ordenando as leituras
temperatura_celsius.sort()
print(f"leituras ordenadas:{temperatura_celsius}")
```

```python
##logica condicional e fluxos de decisao
###exemplo 01 controle de nivel liquido

#simulação de leitura de nivel liquido
nivel_litros = 50 #nivel atual de liquido em litros
nivel_maximo = 100 #nivel maximo permitido em litros
nivel_minimo = 20 #nivel minimo seguro em litros

print (f"Nivel atual do liquido: {nivel_litros} litros")

#logica condicional para controle do nivel liquido
if nivel_litros > nivel_maximo:
  print("Atenção: nivel liquido acima do maximo! Acionando drenagem.")
elif nivel_litros < nivel_minimo:
  print("Atenção: nivel liquido abaixo do minimo! Acionando sistema de abastecimento")
else:
  print("Nivel do liquido dentro dos limites normais.")
```

```python
#exemplo 02 monitoramento de pressão
#simulando valores lidos
pressao_atual = 3.5 #pressão atual em bar
pressao_maxima = 5.0 #pressao maxima permitida em bar
pressao_minima = 1.0 #pressao minima permitida em bar

print (f"Pressão atual: {pressao_atual} bar")

#logica condicional para controle da pressão
if pressao_atual > pressao_maxima and pressao_atual > 4.0:
  print("Atenção pressao critica! acionando alivio de pressão.")
elif pressao_atual < pressao_minima or pressao_atual < 2.0:
  print ("atenção: pressão baixa! acionando compressão.")
else:
  print("Pressão dentro dos limites normais")
```

```python
#exemplo de while
#Temperatura inicial em graus celsius
temperatura = 20

#Enquanto a temperatura for menor ou igual a 25
while temperatura <=25:
  print(f"temperatura atual: {temperatura:.2f}°C")
  temperatura+=0.5 #aumenta a demperatura em 1 grau °c
```

Fluxos Iterativos e Laços de Repetição

```python
##Parte 1 – Loop while: Monitoramento contínuo

#Importando biblioteca para simular tempo
import time

#definindo o consumo de energia inicial
consumo_energia = 50 #consumo inicial em kw
limite_superior = 100 #limite superior de consumo seguro em kw

#laços de repetição para monitorar o consumo de energia
while True:
  print(f"Consumo de energia atual: {consumo_energia} KW")
  if consumo_energia > limite_superior:
    print("Alerta: Consumo de energia acima do limite seguro!")
    break #sai do loop se o limite for ultrapassado
  #simulando variação de energia
  consumo_energia +=10 #aumenta o consumo em 10kw
  time.sleep(2) #Pausa de 2 segundo entre as leituras
print ("Monitoramento encerrado.")
```

```python
##Parte 2 – Loop for: Iterando com range(), lista e dicionário

###Exemplo 1 – Usando for com range()
#O loop irá iterar de 1 a 5, representando cada ciclo de consumo

for ciclo in range(1, 6):
  #calculo do consumo de energia para o ciclo atual
  consumo_energia = 50 + (ciclo *8)
  #Exibe o ciclo e o respectivo consumo de energia em kw
  print(f"Ciclo{ciclo}: Consumo de energia = {consumo_energia} KW")
```

```python
# O loop irá gerar valores de 1 a 9, pulando de 2 em 2
for ciclo in range(1, 10, 2):  # A cada dois valores
    print(ciclo)  # Exibe o valor atual do ciclo
```

```python
#Exemplo 02- usando for e listas
#Lista com dados simulados de consumo de energia
leituras = [55, 63, 70, 95, 105]

#itera sobre cada leitura na lista
for leitura in leituras:
  #verifica se a leitura de consumo excede o limite de 100 kw
  if leitura > 100:
    print(f"Alerta! consumo de {leitura} KW excedeu o limite!")
  else:
    print(f"consumo de {leitura} KW dentro do limite.")
```

```python
##Exemplo 3 – Usando for com dicionário
#dicionario contendo o consumo de energia em diferentes setores
consumo_setores = {
    "Produção": 88,
    "refrigeração":102,
    "Iluminacao":76
}

#Itera sobre cada setor e seu respectivo consumo de energia
for setor, consumo in consumo_setores.items():
  #determina o status do consumo com base no limite de 100 kw
  if consumo >100:
    status = "acima do limite"
  else:
    status = "dentro do limite"
  #exibe o setor, consumo e o status
  print (f"Setor: {setor} | Consumo: {consumo} kw- status: {status}")
```

>Análise da Lógica
Combinamos loops while e for para simular cenários reais de monitoramento e análise de consumo energético:

while: ideal para monitoramento contínuo com condição de parada.
for com range(): útil para repetições com número conhecido de ciclos.
for com lista: para percorrer coleções de dados.
for com dicionário: para analisar dados relacionados (ex: setor x consumo).
